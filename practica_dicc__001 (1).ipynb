___Ejercicio 1:  
Hacer una función que reciba una lista de diccionarios de provincias y un area.  
La función tiene que retornar una lista con las provincias del area.___   

```python
def provincias_areas(lista_provincias: list, area: str) -> list:
    nueva_lista = []    
    for provincia in lista_provincias:
        if provincia['area'] == area:
            nueva_lista.append(provincia)

    return nueva_lista

provincias_areas(provincias_ejemplo, 'Litoral')
```
